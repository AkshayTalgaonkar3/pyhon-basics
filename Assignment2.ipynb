Task III

```python
input_option = '1'
orderList = []

while input_option != 'q':
    input_option= input('enter option')
    if input_option == '1':
        order = int(input('enter order'))
        orderList.append(order)
    elif input_option == '2':
        final_amount_total = 0
        for order in orderList :
            discount = 0
            if order >= 2000:
                discount = (order * 15)/100
            elif order >= 1500 and order < 2000 :
                discount = (order * 10)/100
            elif order >= 1000 and order < 1500 :
                discount = (order * 7)/100
            else:
                discount = 0
            print(discount)
            final_amount = order - discount
            print(final_amount)
            final_amount_total = final_amount_total + final_amount

        print(final_amount_total)

    elif input_option == 'q':
        break;
```
