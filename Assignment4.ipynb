Task VII

```python
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

discount = float(input("Enter discount percentage: "))

filename = "discount_report.txt"

file = open(filename, "w")

file.write("Product | Original Price | Discounted Price\n")

total_discounted_price = 0

for product, price in prices.items():
    discounted_price = price - (price * discount / 100)
    total_discounted_price += discounted_price

    file.write(f"{product} | {price} | {discounted_price:.2f}\n")

# summary (extra)
file.write("\nSummary\n")
file.write(f"Total Items: {len(prices)}\n")
file.write(f"Average Discounted Price: {total_discounted_price / len(prices):.2f}\n")

file.close()

# read and print file content
file = open(filename, "r")
print("\n--- Discount Report ---\n")
print(file.read())
file.close()
```
