```python
price_dict = {'Pen':20,'WaterBottel':50,'Bag':400,'Shoes':700,'Bat':600,'Laptop':35000}

price_dict['Pen'] = 30

print(price_dict)

if('Laptop' in price_dict):
    price_dict.pop('Laptop')
else:
    print('Laptop does not exist')

print(price_dict)

sum = 0
count = 0
for key,values in price_dict.items():

    sum = sum + values
    count = count + 1

avg= sum / count

print(avg)
```
