```python
from abc import ABC, abstractmethod

class Payment:
    @abstractmethod
    def process_payment(self,amount):
        pass

class CreditCardPayment(Payment):
    def process_payment(self,amount):
        print(f'Credit card payment{amount}')

class UPIPayment(Payment):
    def process_payment(self,amount):
        print(f'UPIPayment payment{amount}')

CreditCardPayment1 = CreditCardPayment()
CreditCardPayment1.process_payment(100)

UPIPayment1 = UPIPayment()
UPIPayment1.process_payment(500)
```

```python
class Product:
    def __init__(self,name,price,category):
        self.name = name
        self.price= price
        self.category=category

    def get_info(self):
        print(f'Name : {self.name} Price : {self.price} Category : {self.category}')

    def get_price(self):
        return self.price
    
    def set_price(self,newPrice):
        self.price = newPrice

    def __str__(self):
        return 'Product 1' + self.name 
    
    def __add__(self, other):
        return self.price + other.price
    
product1 = Product('Pen',20,'stationary')
product2 = Product('Notebook',60,'stationary')

product3 = product1 + product2

print(product3)
```

```python
class Product:
    
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # operator overloading
    def __add__(self, other):
        return self.price + other.price


class Inventory:

    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def remove_product(self, name):
        for p in self.products:
            if p.name == name:
                self.products.remove(p)

    def get_total_value(self):
        total = 0
        for p in self.products:
            total += p.price
        return total

    def show_all_products(self):
        for p in self.products:
            print(f"Product: {p.name}, Price: {p.price}")


class Store:

    def __init__(self, store_name):
        self.store_name = store_name
        self.inventory = Inventory()

    def add_new_product(self):

        name = input("Enter product name: ")
        price = float(input("Enter price: "))

        product = Product(name, price)

        self.inventory.add_product(product)

    def show_summary(self):

        print("Store:", self.store_name)

        total_items = len(self.inventory.products)
        total_value = self.inventory.get_total_value()

        print("Total Items:", total_items)
        print("Total Value:", total_value)



store = Store("My Store")

store.add_new_product()
store.add_new_product()
store.add_new_product()

print("\nAll Products:")
store.inventory.show_all_products()

print("\nStore Summary:")
store.show_summary()

# using __add__
p1 = Product("Pen", 10)
p2 = Product("Notebook", 50)

print("\nCombined price using __add__:", p1 + p2)
```
